```python
# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('../dados/trees.csv')
base.head()
```

```python
# girth com volume
plt.scatter(base.Girth, base.Volume) # gráfico de dispersão, onde o eixo x é o girth e o eixo y é o volume
```

```python
# girth com heigth
plt.scatter(base.Girth, base.Height) # gráfico de dispersão, onde o eixo x é o girth e o eixo y é a altura
```

```python
# heigth com volume
plt.scatter(base.Height, base.Volume, marker = '*')
# gráfico de dispersão, onde o eixo x é a altura e o eixo y é o volume e o marcador é um asterisco
```

```python
# hsitograma volume
plt.hist(base.Volume) # histograma do volume
```

```python
#imprimindo juntos
# Criação de figura, no qual os gráficos serão posicionados
plt.figure(1) # cria a figura 1
plt.subplot(2,2,1) # divide a figura em 2 linhas, 2 colunas e plota na posição 1
plt.scatter(base.Girth, base.Volume) # gráfico de dispersão, onde o eixo x é o girth e o eixo y é o volume
plt.subplot(2,2,2) # divide a figura em 2 linhas, 2 colunas e plota na posição 2
plt.scatter(base.Girth, base.Height) # gráfico de dispersão, onde o eixo x é o girth e o eixo y é a altura
plt.subplot(2,2,3) # divide a figura em 2 linhas, 2 colunas e plota na posição 3
plt.scatter(base.Height, base.Volume, marker = '*') # gráfico de dispersão, onde o eixo x é a altura e o eixo y é o volume e o marcador é um asterisco
plt.subplot(2,2,4) # divide a figura em 2 linhas, 2 colunas e plota na posição 4
plt.hist(base.Volume)  # histograma do volume
```
