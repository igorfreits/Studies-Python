```python
# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('../dados/trees.csv')
base.head()
```

```python
# Geração do boxplot
# patch_artist = True preenche, showfliers outliers
plt.boxplot(base.Volume, vert = False, showfliers = False, notch = True,patch_artist = True)
# vert = Posição do boxplot, se fosse True seria na vertical
# showfliers = False não mostra os outliers, valores discrepantes
# notch = Faz a marcação no boxplot na mediana
# patch_artist = True preenche o boxplot

plt.title('Árvores') # título do gráfico
plt.xlabel('Volume') # legenda eixo horizontal
```

```python
#dados por linha
plt.boxplot(base) # gera o boxplot para todas as colunas da base de dados
plt.title('Árvores') # título do gráfico
plt.xlabel('Dados') # legenda eixo horizontal
```

```python
# Geração de 3 boxplots, cada um mostrado informações diferentes
plt.boxplot(base.Volume, vert = False) # boxplot Volume
plt.boxplot(base.Girth, vert = False) # boxplot Girth
plt.boxplot(base.Height, vert = False)  # boxplot Height
plt.title('Árvores') # título do gráfico
plt.xlabel('Dados') # legenda eixo horizontal
```
