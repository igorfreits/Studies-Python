```python
# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
# Carregamento da base de dados
base = pd.read_csv('../dados/orchard.csv')
base
```

```python
# Criação do gráfico 3D, indicando o atributo projection = '3d' e passando três atributos (decrease, rowpos e colpos)
figura = plt.figure() # Cria a figura
eixo = figura.add_subplot(1, 1, 1, projection = '3d')  # Cria o eixo, indicando que é 3D
eixo.scatter(base.decrease, base.rowpos, base.colpos) #  Cria o gráfico 3D
eixo.set_xlabel('decrease')  # Legenda do eixo x
eixo.set_ylabel('rowpos') # Legenda do eixo y
eixo.set_zlabel('colpos') # Legenda do eixo z
```
