```python
# Importação do pandas para leitura de arquivos .csv
import pandas as pd
```

```python
# Carregamento da base de dados
base = pd.read_csv('../dados/insect.csv')
base.shape
```

```python
#dados
base.head() # Exibe as 5 primeiras linhas
```

```python
# Agrupamento dos dados baseado no atributo 'spray', contando e somando os registros
agrupado = base.groupby(['spray'])['count'].sum() # Conta a quantidade de registros
agrupado
```

```python
# Gráfico de barras
agrupado.plot.bar(color = 'gray') # Gera o gráfico de barras, com a cor cinza
```

```python
#cores
agrupado.plot.bar(color = ['blue','yellow','red','green','pink','orange']) # Gera o gráfico de barras, com cores diferentes
```

```python
# Gráfico de pizza
agrupado.plot.pie() # Gera o gráfico de pizza
```

```python
#com legenda
agrupado.plot.pie(legend = True) # Gera o gráfico de pizza, com legenda
```
