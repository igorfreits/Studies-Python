```python
# Importação das bibliotecas
from igraph import Graph
from igraph import plot
```

```python
# recriamos o grafo 4
grafo4 = Graph(directed = False)
grafo4.add_vertices(5)
grafo4.add_edges([(0,1),(1,2),(2,3),(3,4),(4,0),(0,2),(2,1)])
grafo4.add_vertex(5)
grafo4.add_vertex(6)
grafo4.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
grafo4.vs['name'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
```

```python
# Impressão da matriz de adjacência
print(grafo4.get_adjacency()) # imprime a matriz de adjacência
```

```python
# linha
grafo4.get_adjacency()[0,] # imprime a linha 0
```

```python
#linha e coluna
grafo4.get_adjacency()[0,1] # imprime o elemento da linha 0 e coluna 1
```

```python
# Estrutura de repetição para percorrer cada vértice, visualizando o nome e o rótulo
for v in grafo4.vs:
   print(v) 
```

```python
plot(grafo4, bbox=(0,0,300,300)) 
```

```python
# Criação de grafo com pesos entre as relações
grafo5 = Graph(edges = [(0,1),(2,3),(0,2),(0,3)], directed = True) # grafo direcionado
grafo5.vs['label'] = ['Fernando', 'Pedro', 'Jose', 'Antonio'] # rótulos
grafo5.vs['peso'] = [40,30,30,25] # pesos
print(grafo5)
```

```python
# Percorrer os vértices para visualizar os pesos
for v in grafo5.vs:
    print(v)
```

```python
grafo5.vs[0]
```

```python
# Definição do tipo de amizado e do peso das relações
grafo5.es['TipoAmizade'] = ['Amigo', 'Inimigo', 'Inimigo', 'Amigo'] # tipo de amizade
grafo5.es['weight'] = [1,2,1,3] # pesos das relações
print(grafo5)
```

```python
# Percorrer os vértices, tipo de amizade
for e in grafo5.es:
    print(e)
```
